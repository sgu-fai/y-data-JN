```python
import re
r = re.compile('\\d{4}')
def get_year(y_str):
    l = r.findall(y_str) # take the first year
    if len(l) == 0:
        return None
    return int(l[0])

sf['year'] = sf['PublicationYear'].apply(lambda s: get_year(s))
# sf['year']
```

**Question 1:** Write a function that returns the most popular book in each subject. Use it to find the most popular book in
    the _Mystery Fiction_ subject (15pt)

```python
title_subject_sf = []
def most_popular_book_by_subject(subject):
    title_subject_sf = sf['Title', 'Subjects']
    title_subject_sf['subject_list'] = title_subject_sf['Subjects'].apply(lambda s: s.split(","))
    title_subject_sf['subject_list'] = title_subject_sf['subject_list'].apply(lambda l: [subject.strip() for subject in l])
    title_subject_sf = title_subject_sf.remove_column('Subjects')
    title_subject_sf = title_subject_sf.unique() 
    title_subject_sf = title_subject_sf.stack("subject_list", new_column_name="subject") 
    return title_subject_sf[title_subject_sf['subject'] == subject].groupby(['Title'],{'Count': agg.COUNT()}).sort('Count', ascending=False)[0]['Title']
```

```python
%%time
most_popular_book_by_subject('Mystery fiction')
```

**Question 2**: Find the top-10 fiction subjects (5pt), and use Seaborn to visualize the number of items over time for each subject (15pt)
    

```python
%%time
sf2 = sf['BibNum','Subjects','year']
sf2['subject_list'] = sf2['Subjects'].apply(lambda s: s.split(","))
sf2['subject_list'] = sf2['subject_list'].apply(lambda l: [subject.strip() for subject in l])
sf2 = sf2.remove_column('Subjects')
sf2 = sf2.unique() 

sf2 = sf2.stack("subject_list", new_column_name="subject") 

sf3 = sf2[sf2['subject'].apply(lambda s: "fiction" in s or "Fiction" in s)]
sf3 = sf3.groupby(['subject','year'],{'count': agg.COUNT()})
top_subjects = sf3.groupby(['subject'],{'count': agg.SUM('count')}).sort('count', ascending=False)[:10]['subject']
print("Top 10 fiction subjects: ",top_subjects)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%%time
plt.figure(figsize=(10,5))
sns.lineplot(data=sf3.filter_by(top_subjects,'subject').to_dataframe(),x='year',y='count',hue='subject')
```

**Question 3:** Write a function that gets as input a subject name. The function returns the top-20 most _popular_ author for the input subject (15pt).



```python
def most_popular_authors_by_subject(subject):
    title_subject_sf = sf['Author', 'Subjects']
    title_subject_sf['subject_list'] = title_subject_sf['Subjects'].apply(lambda s: s.split(","))
    title_subject_sf['subject_list'] = title_subject_sf['subject_list'].apply(lambda l: [subject.strip() for subject in l])
    title_subject_sf = title_subject_sf.remove_column('Subjects')
    title_subject_sf = title_subject_sf.unique() 
    title_subject_sf = title_subject_sf.stack("subject_list", new_column_name="subject") 
    return title_subject_sf[title_subject_sf['subject'] == subject].groupby(['Author'],{'Count': agg.COUNT()}).sort('Count', ascending=False)[:20]['Author']
```

```python
most_popular_authors_by_subject('Science fiction')
```

**Question 5:** Using the  [SJR Journal Ranking of 2018](https://www.scimagojr.com/journalrank.php) dataset and Seaborn, visualize
    the H-index distributions according to the SJR Best Quartile (15pt). Using a grid, visualize the _Cites / Doc. (2years)_ histograms values of each publisher (15pt).
    


```python
sf = tc.SFrame.read_csv('datasets/journal_ranking/sjr2018.csv',delimiter=';',verbose=False)

def convert_comma_str_to_float(s):
    try:
        return float(s.replace(",", "."))
    except:
        return 0
    
for i in ["SJR", "Cites / Doc. (2years)", "Ref. / Doc."]:
    sf[i] = sf[i].apply(lambda s: convert_comma_str_to_float(s))
    
df = sf.to_dataframe()
```

```python
plt.figure(figsize=(10,10))
g = sns.FacetGrid(df, hue='SJR Best Quartile',height=5) 
g.map(sns.distplot, 'H index',).add_legend()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=df,y='H index', x='SJR Best Quartile') 
```
